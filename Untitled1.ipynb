```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# 그리드 생성
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

# potential flow 함수 정의
def potential_flow_source_sink(X, Y, strength_source, x_source, y_source, strength_sink, x_sink, y_sink):
    """
    X, Y: 그리드 포인트의 좌표
    strength_source, x_source, y_source: 싱크의 강도 및 위치
    strength_sink, x_sink, y_sink: 싱크의 강도 및 위치
    """
    # 소스 및 싱크의 속도 구하기
    u_source = strength_source / (2 * np.pi) * (X - x_source) / ((X - x_source)**2 + (Y - y_source)**2)
    v_source = strength_source / (2 * np.pi) * (Y - y_source) / ((X - x_source)**2 + (Y - y_source)**2)

    u_sink = strength_sink / (2 * np.pi) * (X - x_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)
    v_sink = strength_sink / (2 * np.pi) * (Y - y_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)

    # 전체 속도 구하기
    u = u_source + u_sink
    v = v_source + v_sink

    return u, v
```

```python

# 소스 및 싱크의 매개 변수 설정
strength_source = 1.0
x_source, y_source = -1.0, 0.0

strength_sink = -1.0
x_sink, y_sink = 1.0, 0.0
```

```python

# potential flow 계산
u, v = potential_flow_source_sink(x, y, strength_source, x_source, y_source, strength_sink, x_sink, y_sink)

# 속도 벡터를 시각화
plt.figure(figsize=(8, 6))
plt.streamplot(x, y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], color='red', marker='o')
plt.title('Potential Flow with Source and Sink')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 그리드 생성
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

# 소용돌이(flow) 함수 정의
def vortex_flow(X, Y, strength_vortex, x_vortex, y_vortex):
    """
    X, Y: 그리드 포인트의 좌표
    strength_vortex, x_vortex, y_vortex: 소용돌이의 강도 및 위치
    """
    # 소용돌이의 속도 구하기
    u_vortex = strength_vortex / (2 * np.pi) * (Y - y_vortex) / ((X - x_vortex)**2 + (Y - y_vortex)**2)
    v_vortex = -strength_vortex / (2 * np.pi) * (X - x_vortex) / ((X - x_vortex)**2 + (Y - y_vortex)**2)

    return u_vortex, v_vortex

# 소용돌이의 매개 변수 설정
strength_vortex = 1.0
x_vortex, y_vortex = 0.0, 0.0

# 소용돌이(flow) 계산
u_vortex, v_vortex = vortex_flow(x, y, strength_vortex, x_vortex, y_vortex)

# 속도 벡터를 시각화
plt.figure(figsize=(8, 6))
plt.streamplot(x, y, u_vortex, v_vortex, density=1, linewidth=1, arrowsize=1, arrowstyle='->')
plt.scatter(x_vortex, y_vortex, color='red', marker='o')
plt.title('Vortex Flow')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 그리드 생성
x, y = np.meshgrid(np.linspace(-3, 3, 400), np.linspace(-3, 3, 400))

# 원의 매개 변수 설정
radius = 1.0
center = (0, 0)

# potential flow 함수 정의
def potential_flow_around_circle(X, Y, strength_source, x_source, y_source, strength_sink, x_sink, y_sink):
    r_source = np.sqrt((X - x_source)**2 + (Y - y_source)**2)
    theta_source = np.arctan2(Y - y_source, X - x_source)

    r_sink = np.sqrt((X - x_sink)**2 + (Y - y_sink)**2)
    theta_sink = np.arctan2(Y - y_sink, X - x_sink)

    u_source = strength_source / (2 * np.pi) * np.cos(theta_source) / r_source
    v_source = strength_source / (2 * np.pi) * np.sin(theta_source) / r_source

    u_sink = strength_sink / (2 * np.pi) * np.cos(theta_sink) / r_sink
    v_sink = strength_sink / (2 * np.pi) * np.sin(theta_sink) / r_sink

    u = u_source + u_sink
    v = v_source + v_sink

    return u, v

# 원 주위의 potential flow 계산
strength_source = 1.0
x_source, y_source = center[0] + radius, center[1]

strength_sink = 1.0
x_sink, y_sink = center[0] - radius, center[1]

u, v = potential_flow_around_circle(x, y, strength_source, x_source, y_source, strength_sink, x_sink, y_sink)

# 속도 벡터를 시각화
plt.figure(figsize=(8, 6))
plt.streamplot(x, y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
plt.title('Potential Flow Around a Circle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(center[0], center[1], color='red', marker='o')  # 원의 중심 표시
plt.show()
```
